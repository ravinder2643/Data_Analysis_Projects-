```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
warnings.filterwarnings("ignore")
hr= pd.read_csv("HR_Analysis.csv",header=0)
hr.head()
```

```python
hr.shape
```

```python
fig,ax= plt.subplots(2,2, figsize=(10,10)) 
plt.suptitle("Understanding the distribution of age", fontsize= 20)
sns.distplot(hr['Age'], ax= ax[0,0])  #plot on 1st axis
ax[0,0].set_title('Distribution of Age')

sns.distplot(hr['TotalWorkingYears'], ax= ax[0,1])  #plot on 2nd axis
ax[0,0].set_title('Distribution of Total Working years')

sns.distplot(hr['YearsAtCompany'], ax= ax[1,0])  #plot on 2nd axis
ax[0,0].set_title('Distribution of years at Company')

sns.distplot(hr['YearsInCurrentRole'], ax= ax[1,1])  #plot on 2nd axis
ax[0,0].set_title('Distribution of years in Current Role')
```

**From above plots, we find that majority of the employees are in age group of 35-40 years and have a total experience of around 8-10 years. They spent around 2-4 years at current organisation and 2-3 years in current role**

***COUNT PLOT***

```python
sns.countplot(data= hr, x= 'Attrition')
plt.show()
```

Above plot show that around 250 people have left organisation while the organisation has been able to retain around 1200 people

***BAR PLOT***

```python
from numpy import median
sns.barplot(data= hr, x= 'Attrition', y= 'MonthlyIncome', hue='Gender', estimator= median)
plt.show()
```

```python
 #BAr plot of Job Satisfaction  with respect to Monthy income from home according to gender
from numpy import median
sns.barplot(data= hr, x= 'Attrition', y='MonthlyIncome', hue ='JobSatisfaction', estimator= median)
plt.show()
```

```python
from numpy import median
sns.barplot(data= hr, x= 'Attrition', y='MonthlyIncome', hue ='MaritalStatus', estimator= median)
plt.show()
```

***Box plot***

```python
f, ax= plt.subplots(figsize= (15, 10))
sns.boxplot(data = hr, x= 'JobSatisfaction', y = 'MonthlyIncome', hue= 'JobLevel', palette= 'Set3')
plt.legend(loc= 'best')
plt.show()
```

```python
sns.violinplot( data =hr, x="Attrition", y = "YearsAtCompany", hue='Gender', split = True, inner= "stick", palette= 'muted')
#setting split true will draw half voilin for each level
```

```python
# jointt scatter plot
sns.jointplot(x='Age', y='MonthlyIncome', data=hr, kind='scatter')
plt.show()
```
